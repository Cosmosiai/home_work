```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(-4*np.pi, 4*np.pi, 100)
xlab='время сек'
ylab='амплитуда сигнала'
y1= np.sin(x)
leg1='гармоничная функция'
y2=10*np.sin(0.5*x)+np.cos(10*x)
leg2='смешанный сигнал'
y3=np.cos(x)
leg3='опорный сигнал'
y4 = x**2/5
leg4='квадратичная функция'
plt.grid(color = 'b',linewidth = 1)
plt.xlabel(xlab, fontsize = 'x-large')
plt.ylabel(ylab, fontsize = 'x-large')
plt.title('Гармонические сигналы ')
plt.plot(x,y1,'b-',label=leg1)
plt.plot(x,y4,'r--',label=leg4)
plt.legend(loc='best')
plt.grid(True)
plt.savefig('signal3.png', format = 'png')
plt.show()
```

```python
t = np.linspace(0, 0.4*np.pi, 1000)
x = np.cos(3*t)
y = np.sin(2 * t)

plt.figure(figsize=(8, 6))
plt.plot(x, y, label='Параметрическая функция: (cos(t), sin(2t))')
plt.title('График параметрической функции')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()
plt.show()
```

```python
years = [2016, 2017,2018, 2019, 2020,2021, 2022]
minimal_live = [4794.34, 4900.79,4792.54, 4806.32,5358.58, 6268.58, 7178.32]

# Постройте столбиковую диаграмму
plt.figure(figsize=(8, 6))
plt.bar(years, minimal_live, color='blue')
plt.title('Прожиточный минимум в Кыргызстане по годам')
plt.xlabel('Год')
plt.ylabel('Прожиточный минимум (сомы)')
plt.grid(axis='y')
plt.show()
```

```python
dpi = 90 
fig = plt.figure(dpi = dpi, figsize = (512 / dpi, 384 / dpi) )
plt.rcParams.update({'font.size': 9})
plt.title('Распределение по годам (%)')
xs = range(len(years))
plt.pie(minimal_live, autopct='%.1f', radius = 1.1,explode = [0.15] + [0 for _ in range(len(years) - 1)] )
# labels = [f'{year} ({value:.1f})' for year, value in zip(years, minimal_live)]
labels = [f'{year}' for year in years]

for i in range(len(years)):
    plt.legend(bbox_to_anchor = (-0.2, 0.50, 0.25, 0.25),loc = 'lower left', labels = labels )
plt.show()
```
